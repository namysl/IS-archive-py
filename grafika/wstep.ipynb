```python
#funkcja do napisania
#zadanie dodatkowe - prosze zmodyfikować funkcje tak, by można było przekazać kolor rysowanego prostokata jako param

def prostokat(x, y, w, h, background_color, color):
    #assert isinstance(background_color, list)
    #funkcja zwraca tablicę RGB typu uint8 o wymiarach 100x100x3
    
    data=np.zeros((100, 100, 3), dtype=np.uint8)
    data.fill(background_color)
    
    for i in range(y, w+x):
        for j in range(x, h+x):
            data[i, j, 0]=color[0]
            data[i, j, 1]=color[1]
            data[i, j, 2]=color[2]
    
    plt.imshow(data, interpolation='none', aspect=1)
    plt.show()
    return data
    
prostokat(50, 50, 20, 20, 0, [0, 206, 209])
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def liniaN(x1, y1, x2, y2, rysowac=False, kolor_linii=[0, 0, 0]):
    assert isinstance(kolor_linii, list)
    assert x1 < 100
    assert x2 < 100
    assert y1 < 100
    assert y2 < 100
    
    if x1 > x2:
        x1, x2 = x2, x1
        y1, y2 = y2, y1
    
    
    data=np.zeros((100, 100, 3), dtype=np.uint8)
    data.fill(255)
    
    delta_y = (y2 - y1)
    delta_x = (x2 - x1)
    
    m = delta_y / delta_x
    print('m =', m)
    
    c = y1

    for x in range(x1, x2):
        y = math.floor(m * (x - x1) + c)
        data[x, y, 0] = kolor_linii[0]
        data[x, y, 1] = kolor_linii[1]
        data[x, y, 2] = kolor_linii[2]
    
    if rysowac is True:
        plt.imshow(data, interpolation='none', aspect=1)
        plt.show()
    
    return data
    

#liniaN(50, 50, 2, 2, True, [0, 206, 209])
liniaN(2, 2, 50, 50)
liniaN(30, 60, 80, 20)
liniaN(50, 50, 30, 30)
liniaN(15, 80, 60, 10)
liniaN(1, 99, 20, 20)
```
