```python
import numpy as np
import matplotlib.pyplot as plt

def Newton( n, k ):
 
    val = 1
    for i in range( 1, k+1 ):
        val = val * ( n - i + 1 ) / i
    return val

def Bernstein(n, i, t):
    B=Newton(n, i)*(t**i)*((1.0-t)**(n-i))
    return B

def pBezier(points):
    data=[]
    data.append([])
    data.append([])
    n = len(points)-1
    ll = 50
    for i in range( ll+1):
        t=i/ll
        
        px=0
        py=0
        for j in range(n+1):
            bb=Bernstein(n, j,t)
            #print(bb)
            px+=points[j][0]*bb
            py+=points[j][1]*bb
        #print(px, py)
        data[0].append(px)
        data[1].append(py)        
    return data
```

```python
dataS = pBezier([[0, 40], [5, 0], [10, 60], [5, 50], [0, 60], [0, 100], [5, 80]])
a=plt.plot(dataS[0], dataS[1])

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches


# define the geometry of a path:
verts = [(0.0, 0.0), # P0
         (0.3, 1.1), # P1
         (1.0, 0.8), # P2
         (0.8, 0.0), # P3
        ]


codes = [Path.MOVETO, # put the "pen" at P0
         Path.CURVE4, # and draw bezier curves from P0 to P1,
         Path.CURVE4, # P1 to P2
         Path.CURVE4, # P2 to P3.
        ]


codes2 = [Path.MOVETO, # put the "pen" at P0
         Path.CURVE3, # and draw bezier curves from P0 to P1,
         Path.CURVE3, # P1 to P2
         Path.CURVE3, # P2 to P3.
        ]




path = Path(verts, codes) # build path from vertices & command codes
path2 = Path(verts, codes2) # build path from vertices & command codes
patch = patches.PathPatch(path, facecolor='none', lw=2)
patch2 = patches.PathPatch(path, facecolor='none', lw=2)
ax = plt.gca()
ax.add_patch(patch)
#ax.add_patch(patch2)




xs, ys = zip(*verts)
ax.plot(xs, ys, 'ro')
plt.show()


ax = plt.gca()
#ax.add_patch(patch)
ax.add_patch(patch2)


ax.plot(xs, ys, 'ro')
plt.show()
```
