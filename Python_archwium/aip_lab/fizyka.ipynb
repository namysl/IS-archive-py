```python
import matplotlib.pyplot as plt 
import math as m

def krzywol(hs=100, a=40, v0=100, g=10):
    plt.xlabel('m') 
    plt.ylabel('m') 
    a = m.radians(a)
    
    t = 1
    osx = [0]
    osy = [hs]
    hmax = hs
    
    #rownania ruchu
    vx = v0*m.cos(a)
    vy = v0*m.sin(a)
    
    while(osy[-1] >= 0):
        x = t
        y = t*m.sin(a)/m.cos(a) - 1/2*g*t*t/(v0*v0*m.cos(a)*m.cos(a))+hs
        osx.append(x)
        osy.append(y)
        
        if(hmax<y):
            hmax = y
        t += 0.25
        
    tw = vy/g
    ts = m.sqrt(((2*hmax)/g))
    tc = tw+ts
    
    zw = vx*tw #zasieg wznoszenia
    zs = vx*ts #zasieg spadania
    zc = zw+zs #zasieg calkowity
    
    plt.plot(osx,osy)
    print("Maksymalna wysokość:", hmax)
    print("Zasięg: ", zc)
    print("Czas wznoszenia się: ", tw)
    print("Czas spadania: ", ts)
    print("Całkowity czas lotu: ", tc)
krzywol()
```

```python
krzywol(50, 50, 10, 10)
```

```python
krzywol(50, 0, 20, 10)
```
